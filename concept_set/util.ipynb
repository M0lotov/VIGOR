```python
from glob import glob

all_concepts = []
for path in glob('output/05_*.txt'):
    with open(path, 'r') as f:
        content = f.readlines()
    all_concepts.extend([line.strip() for line in content if line.strip()])
all_concepts = sorted(list(set(all_concepts)))

with open('output/05_all_concepts.txt', 'w') as f:
    for concept in all_concepts:
        f.write(f"{concept}\n")
```

```python
from collections import defaultdict
from glob import glob

for cat in ['color', 'edge', 'texture', 'shape', 'part', 'object', 'motion', 'relation']:
    final_concepts = []
    concept_count = defaultdict(int)
    for path in glob(f'output/*/08_filtered_by_category_txt/{cat}.txt'):
        with open(path, 'r') as f:
            concepts = f.readlines()
            for concept in concepts:
                concept_count[concept.strip()] += 1
    for concept in concept_count:
        if concept_count[concept] > 1:
            final_concepts.append(concept)
    with open(f'output/final_concepts/{cat}.txt', 'w') as f:
        for concept in sorted(final_concepts):
            f.write(f'{concept}\n')
    print(f'{cat}: {len(concepts)} -> {len(final_concepts)}')
```
